Given a matrix ofÂ mÂ xÂ n. Write a `transpose()` function that will return the transpose matrix n x m.

ðŸ‘€ Example

Input
```
1 2
3 4
5 6
```

Output
```

```

````python
const transpose = (matrix) => {
    let ans = []
    for (let row = 0; row < matrix[0].length; row++) {
        let temp = []
        for (let col = 0; col < matrix.length; col++) {
            temp.push(matrix[col][row]);
        }
        ans.push(temp)
    }
    console.log(ans)
}
transpose([[1,2],[3,4],[5,6]])
````

â“ **Spiral order**

Given a matrix ofÂ nÂ xÂ nÂ elements, return all elements of the matrix in spiral order.

ðŸ‘€ Example

Input 
```
 1 2 3
 4 5 6 
 7 8 9
```

output
```
1 2 3 6 9 8 7 4 5
```

````python

const spiralTransversal = (mat) => {
    let ans = []
    let left = 0;
    let right = mat[0].length-1;
    let up = 0;
    let down = mat.length-1;
    while (left <= right && up <= down ) {

    //  store ans for remaining upper row
        for (let i = left; i <= right ; i++) {
            ans.push(mat[up][i])
        }

        up += 1
    
    //  store ans for remaining right column
        for (let i = up; i <= down ; i++) {
            ans.push(mat[i][right])
        }
        right -= 1

    //  store ans for remaining down rows
        if (up < down){
            for (let i = right; i >= left ; i--) {
                ans.push(mat[down][i])
            }
        
            down -= 1
        }
    //  store ans for remaining left col
        if (left < right){
            for (let i = down; i >= up; i--) {
                ans.push(mat[i][left])   
            }
            left += 1
        }
    }
    console.log(ans)
}

spiralTransversal([[1,2,3],[4,5,6],[7,8,9]])
````
