â“ Extend the above problem by doing inverse casing instead i.e. output should be `["aPpLE", "BANANA", "oRANGe"]`

````python
lst = ["ApPle","banana","OrangE"]
newlst = []
for word in lst:
    tempword= ''
    for letter in word:
        if(letter.isupper()):
           tempword += letter.lower()
        else:
           tempword += letter.upper()
    newlst.append(tempword)
print(newlst)  
````

â“ Given an integer `n`, return the length of the longest consecutive run of 1s in its binary representation.
For example, given `156`, you should return `3`.

Explanation : `156` in binary from `10011100`, so since `111` is the longest consecutive run, return `3`

````python
def findLongestConsecutive(num):
    ans = bin(num)
    ans = ans.replace('0b','')
    count = 0
    maxcount = 0
    for i in range(len(ans)):
        if(ans[i] == '1'):
            count += 1
        else:
            count = 0
        maxcount = max(count , maxcount)
    return maxcount

findLongestConsecutive(156)

    
    
    
````

â“Given an `array` and a number `k` that's smaller than the length of the array, rotate the array to the right `k` elements in-place. \
For example ```[1,2,3,4,5,6] ``` and `k = 2` we should rotate the array to 2 positions. output : ``` [5,6,1,2,3,4] ```

````python
def rotate(array , k):
    n = len(array)
    if(k>n ):
        return 'K should be smaller then length of array'
    if (k < 0):
        return 'K should not be negative'
         
    array = array[n-k:n] + array[:n-k]
    return array

rotate([1,2,3,4,5,6],2)

````
