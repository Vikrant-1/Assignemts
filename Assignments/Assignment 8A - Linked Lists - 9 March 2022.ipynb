### ‚ùì [#5] Reverse the Linked List

Example: Given Linked list 1->2->3->4, after reversing the List should look like 4->3->2->1

```python
class LinkedList extends BaseLinkedList{
    
    constructor(){
        this.head = new Node(10);
        this.head.next = new Node(20);
        this.head.next.next = new Node(30);
        this.head.next.next.next = new Node(40);
        this.head.next.next.next.next = new Node(50);
    }

    reverseLinkedList(){
        let prev = null;
        let current = this.head;
        while (current) {
            let next = current.next;
            current.next = prev;
            prev = current;
            current = next;
        }
        this.head = prev;
    }

}


let sll = LinkedList()
sll.reverseLinkedList()

console.assert(sll.parse() == [50, 40, 30, 20, 10])
```
